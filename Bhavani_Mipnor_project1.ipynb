```python
# Import NumPy

import numpy as np
```

```python
with open("hostel_bois.txt","r",encoding="utf-8") as file:
    chat = file.readlines()

print("File loaded successfully")
print("Total lines :",len(chat))
```

```python
# Display first few lines

for i in range(10):
    print(chat[i])
```

```python
# Extract Date, Time, Sender and Message
# Using only string methods

messages = []

for line in chat:

    line = line.strip()

    if " - " not in line:
        continue

    first = line.split(" - ",1)

    if len(first) != 2:
        continue

    date_time = first[0]
    remaining = first[1]

    if ": " not in remaining:
        continue

    second = remaining.split(": ",1)

    sender = second[0]
    message = second[1]

    if "," in date_time:

        temp = date_time.split(",")

        date = temp[0].strip()
        time = temp[1].strip()

        messages.append([date,time,sender,message])

print("Messages Extracted :",len(messages))
```

```python
# Find total participants

participants = set()

for data in messages:
    participants.add(data[2])

print("Total Participants :", len(participants))

print("\nParticipants are:")

for person in participants:
    print(person)
```

```python
# Count messages of each participant

message_count = {}

for data in messages:

    sender = data[2]

    if sender in message_count:
        message_count[sender] = message_count[sender] + 1
    else:
        message_count[sender] = 1

print("Messages sent by each participant:\n")

for person in message_count:
    print(person, ":", message_count[person])
```

```python
# Find most active and least active member

highest = 0
lowest = len(messages)

most_active = ""
least_active = ""

for person in message_count:

    if message_count[person] > highest:
        highest = message_count[person]
        most_active = person

    if message_count[person] < lowest:
        lowest = message_count[person]
        least_active = person

print("Most Active Member :", most_active)
print("Messages :", highest)

print()

print("Least Active Member :", least_active)
print("Messages :", lowest)
```

```python
# Calculate average message length

total = 0

for data in messages:
    total = total + len(data[3])

average = total / len(messages)

print("Average Message Length :", round(average,2))
```

```python
# Find most common words

word_count = {}

for data in messages:

    sentence = data[3].lower()

    words = sentence.split()

    for word in words:

        if word in word_count:
            word_count[word] = word_count[word] + 1
        else:
            word_count[word] = 1

print("Top 10 Most Common Words\n")

for i in range(10):

    highest = 0
    common_word = ""

    for word in word_count:

        if word_count[word] > highest:
            highest = word_count[word]
            common_word = word

    print(common_word, ":", highest)

    word_count[common_word] = -1
```

```python
# Count media messages

media = 0

for data in messages:

    if "<Media omitted>" in data[3]:
        media = media + 1

print("Total Media Shared :", media)
```

```python
# Count deleted messages

deleted = 0

for data in messages:

    if "This message was deleted" in data[3]:
        deleted = deleted + 1

print("Deleted Messages :", deleted)
```

```python
# Count system messages

system = 0

for line in chat:

    line = line.strip()

    if " - " in line:

        temp = line.split(" - ",1)

        if ": " not in temp[1]:
            system = system + 1

print("System Messages :", system)
```

```python
# Find most active day

day_count = {}

for data in messages:

    day = data[0]

    if day in day_count:
        day_count[day] = day_count[day] + 1
    else:
        day_count[day] = 1

highest = 0
best_day = ""

for day in day_count:

    if day_count[day] > highest:
        highest = day_count[day]
        best_day = day

print("Most Active Day :", best_day)
print("Messages :", highest)
```

```python
# Find most active hour

hour_count = {}

for data in messages:

    hour = data[1].split(":")[0]

    if hour in hour_count:
        hour_count[hour] = hour_count[hour] + 1
    else:
        hour_count[hour] = 1

highest = 0
best_hour = ""

for hour in hour_count:

    if hour_count[hour] > highest:
        highest = hour_count[hour]
        best_hour = hour

print("Most Active Hour :", best_hour + ":00")
print("Messages :", highest)
```

```python
# Activity Heatmap

heatmap = np.zeros((24,7),dtype=int)

for data in messages:

    hour = int(data[1].split(":")[0])

    day = int(data[0].split("/")[0]) % 7

    heatmap[hour][day] = heatmap[hour][day] + 1

print("Activity Heatmap")

print(heatmap)
```

```python
# Personality of participants

print("Personality Archetypes\n")

for person in message_count:

    count = message_count[person]

    if count >= 700:
        personality = "Spammer"

    elif count >= 500:
        personality = "Conversationalist"

    elif count >= 250:
        personality = "Active Member"

    elif count >= 100:
        personality = "Balanced User"

    else:
        personality = "Ghost Member"

    print(person, ":", personality)
```

```python
# Final Report

print("\n========== GROUPDNA REPORT ==========\n")

print("Total Messages :", len(messages))

print("Total Participants :", len(participants))

print("Most Active Member :", most_active)

print("Least Active Member :", least_active)

print("Average Message Length :", round(average,2))

print("Media Shared :", media)

print("Deleted Messages :", deleted)

print("System Messages :", system)

print("Most Active Day :", best_day)

print("Most Active Hour :", best_hour + ":00")

print("\nMessages Sent By Participants\n")

for person in message_count:
    print(person, ":", message_count[person])

print("\n========== END OF REPORT ==========")
```
